```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

# Display data frame to ensure that it uploaded properly
purchase_data.head()
```

```python
# Show column headers to get proper names for accessing data
purchase_data.columns
```

## Player Count

```python
# Get the number of unique serial numbers to find the total number of players 
nPlayers = purchase_data["SN"].nunique()
print("Player Count:", nPlayers)
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Count the number players by gender

# Create a data frame with only unique players
playerCountDF = purchase_data.drop_duplicates(subset=["SN"])
# Make the data frame with only the unique gender data
playerCountDF = playerCountDF[["SN", "Gender"]]
# Rename SN to Total Count for cleaner output
playerCountDF = playerCountDF.rename(columns={"SN": "Total Count"})
# Get the count of genders
playerCountDF = playerCountDF.groupby(["Gender"]).count()
playerCountDF
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Create a new data frame with the purchase ammount by gender
purchaseAnalysisCount = purchase_data.groupby(["Gender"]).count()
# Clean up columns for output
purchaseAnalysisCount = purchaseAnalysisCount.rename(columns={"Purchase ID": "Purchase Count"})
purchaseAnalysisCount = purchaseAnalysisCount[["Purchase Count"]]
purchaseAnalysisCount.head()
```

```python
# Create a new data frame with the average purchase price by gender
purchaseAnalysisAvePrice = purchase_data.groupby(["Gender"]).mean()
# Clean up columns for output
purchaseAnalysisAvePrice = purchaseAnalysisAvePrice.rename(columns={"Price": "Average Purchase Price"})
purchaseAnalysisAvePrice = purchaseAnalysisAvePrice[["Average Purchase Price"]]
purchaseAnalysisAvePrice["Average Purchase Price"] = purchaseAnalysisAvePrice["Average Purchase Price"].map("${:.2f}".format)
purchaseAnalysisAvePrice
```

```python
# Create a new data frame with the total purchase value by gender
purchaseAnalysisTotalValue = purchase_data.groupby(["Gender"]).sum()
# Clean up columns for output
purchaseAnalysisTotalValue = purchaseAnalysisTotalValue.rename(columns={"Price": "Total Purchase Value"})
purchaseAnalysisTotalValue = purchaseAnalysisTotalValue[["Total Purchase Value"]]
purchaseAnalysisTotalValue["Total Purchase Value"]= purchaseAnalysisTotalValue["Total Purchase Value"].map("${:.2f}".format)
purchaseAnalysisTotalValue
```

```python
# Create data frame for 

# Create a data frame with each players total purchase
purchaseAnalysisAveTotal = purchase_data.groupby(["SN"]).sum()
# Reset the index
purchaseAnalysisAveTotal = purchaseAnalysisAveTotal.reset_index()
# Merge with original data frame to add gender column back in
purchaseAnalysisAveTotal = pd.merge(purchaseAnalysisAveTotal, purchase_data, on="SN", how="inner")
# Drop the duplicate entrys created by the merge
purchaseAnalysisAveTotal = purchaseAnalysisAveTotal.drop_duplicates(subset=["SN"])
# Group the data frame by gender and calculate average of price
purchaseAnalysisAveTotal = purchaseAnalysisAveTotal.groupby(["Gender"]).mean()
# Clean up columns for output
purchaseAnalysisAveTotal = purchaseAnalysisAveTotal[["Price_x"]]
purchaseAnalysisAveTotal = purchaseAnalysisAveTotal.rename(columns={"Price_x": "Avg Total Purchase per Person"})
purchaseAnalysisAveTotal["Avg Total Purchase per Person"] = purchaseAnalysisAveTotal["Avg Total Purchase per Person"].map("${:.2f}".format)
purchaseAnalysisAveTotal
```

```python
# Merge data frames to create final result data frame
purchaseAnalysisCount = pd.merge(purchaseAnalysisCount, purchaseAnalysisAvePrice, on="Gender")
purchaseAnalysisCount = pd.merge(purchaseAnalysisCount, purchaseAnalysisTotalValue, on="Gender")
purchaseAnalysisCount = pd.merge(purchaseAnalysisCount, purchaseAnalysisAveTotal, on="Gender")

# Clean up columns
purchaseAnalysisCount = purchaseAnalysisCount[["Purchase Count", "Average Purchase Price", 
                                               "Total Purchase Value", "Avg Total Purchase per Person"]]

# Display final results data frame
purchaseAnalysisCount
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create the bins to sort the data based on age
bins = [0, 9, 14, 19, 24, 29, 34, 39, 500]

# Create labels for the bins
groupNames = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
# Create the data frame to hold the results
ageDF = purchase_data.drop_duplicates(subset=["SN"])
ageDF = ageDF[["Age"]]
# Sort players into bins based on age
ageDF["Age Ranges"] = pd.cut(ageDF["Age"], bins, labels=groupNames, include_lowest=True)
# Get the total number of players per category
ageDF = ageDF.groupby("Age Ranges").count()

ageDF = ageDF.rename(columns={"Age": "Total Count"})
ageDF
```

```python
# Calculate Percentage of Players for each bin and add the column to the data frame
ageDF["Percentage of Players"] = 100 * ageDF["Total Count"] / nPlayers

# Clean up format for output
ageDF["Percentage of Players"] = ageDF["Percentage of Players"].map("{:.2f}%".format)
ageDF
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Create a data to work from with age bins
ageAnalysis = purchase_data

# Sort players into bins based on age
ageAnalysis["Age Ranges"] = pd.cut(ageAnalysis["Age"], bins, labels=groupNames, include_lowest=True)
ageAnalysis
```

```python
# Calulate Purchase Count
purchaseAgeCount = ageAnalysis.groupby("Age Ranges").count()

# Clean up data frame
purchaseAgeCount = purchaseAgeCount[["Purchase ID"]]
purchaseAgeCount = purchaseAgeCount.rename(columns={"Purchase ID": "Purchase Count"})
purchaseAgeCount
```

```python
# Calulcate Average Purchase Price
purchaseAvePrice = ageAnalysis.groupby("Age Ranges").mean()

# Clean up data frame output
purchaseAvePrice = purchaseAvePrice[["Price"]]
purchaseAvePrice = purchaseAvePrice.rename(columns={"Price": "Average Purchase Price"})
purchaseAvePrice["Average Purchase Price"] = purchaseAvePrice["Average Purchase Price"].map("${:.2f}".format)
purchaseAvePrice
```

```python
# Calulate Total Purchase Value
purchaseTotalValue = ageAnalysis.groupby("Age Ranges").sum()

# Clean up data frame for output
purchaseTotalValue = purchaseTotalValue[["Price"]]
purchaseTotalValue = purchaseTotalValue.rename(columns={"Price": "Total Purchase Value"})
purchaseTotalValue["Total Purchase Value"] = purchaseTotalValue["Total Purchase Value"].map("${:,.2f}".format)
purchaseTotalValue
```

```python
# Calulate Average Total Purchase per Person

# Calculate sum of purchase for each person
purchaseAvePerson = ageAnalysis.groupby("SN").sum()

# Merge data frames to apply bin labels
purchaseAvePerson = pd.merge(purchaseAvePerson, ageAnalysis, on="SN")
# Remove duplicates
purchaseAvePerson = purchaseAvePerson.drop_duplicates(subset=["SN"])

# Group based on age ranges and calulate average
purchaseAvePerson = purchaseAvePerson.groupby("Age Ranges").mean()

# Clean up data frame for output
purchaseAvePerson = purchaseAvePerson[["Price_x"]]
purchaseAvePerson = purchaseAvePerson.rename(columns={"Price_x": "Avg Total Purchase per Person"})
purchaseAvePerson["Avg Total Purchase per Person"] = purchaseAvePerson["Avg Total Purchase per Person"].map("${:.2f}".format)
purchaseAvePerson
```

```python
# Merge results for final data frame

ageAnalysisResults = pd.merge(purchaseAgeCount, purchaseAvePrice, on="Age Ranges")
ageAnalysisResults = pd.merge(ageAnalysisResults, purchaseTotalValue, on="Age Ranges")
ageAnalysisResults = pd.merge(ageAnalysisResults, purchaseAvePerson, on="Age Ranges")

ageAnalysisResults
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Get the purchase count of each person
pc = purchase_data.groupby("SN").count()

# Clean up
pc = pc[["Purchase ID"]]
pc = pc.rename(columns={"Purchase ID": "Purchase Count"})

pc
```

```python
# Get the Average Purchase Price of each person
pp = purchase_data.groupby("SN").mean()

# Clean Up
pp = pp[["Price"]]
pp = pp.rename(columns={"Price": "Average Purchase Price"})
pp["Average Purchase Price"] = pp["Average Purchase Price"].map("${:.2f}".format)

pp
```

```python
# Get the total purchase value
pv = purchase_data.groupby("SN").sum()

# Clean up
pv = pv[["Price"]]
pv = pv.rename(columns={"Price": "Total Purchase Value"})

pv
```

```python
# Merge data frames for output
topSpendersSummary = pd.merge(pc, pp, on="SN")
topSpendersSummary = pd.merge(topSpendersSummary, pv, on="SN")

# Sort by descending Total Purchase Value
topSpendersSummary = topSpendersSummary.sort_values(by=["Total Purchase Value"], ascending=False)

# Clean up format
topSpendersSummary["Total Purchase Value"] = topSpendersSummary["Total Purchase Value"].map("${:.2f}".format)

topSpendersSummary.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Calculate Average Item Price

aveItemPrice = purchase_data[["Item ID", "Item Name", "Price"]]
aveItemPrice = aveItemPrice.groupby(["Item ID", "Item Name"]).mean()
aveItemPrice = aveItemPrice.rename(columns={"Price": "Item Price"})

aveItemPrice.head()
```

```python
# Get Purchase Count
purchaseCount = purchase_data[["Item ID", "Item Name", "Price"]]
purchaseCount = purchaseCount.groupby(["Item ID", "Item Name"]).count()
purchaseCount = purchaseCount.rename(columns={"Price": "Purchase Count"})

purchaseCount.head()
```

```python
# Get Total Purchase Value
tpv = purchase_data[["Item ID", "Item Name", "Price"]]
tpv = tpv.groupby(["Item ID", "Item Name"]).sum()
tpv = tpv.rename(columns={"Price": "Total Purchase Value"})

tpv.head()
```

```python
# Merge data frames for results
popularItems = purchaseCount.join(aveItemPrice, how="outer")
popularItems = popularItems.join(tpv, how="outer")

# Sort in descending order of Total Purchase Value
popularItemsSorted = popularItems.sort_values(by=["Purchase Count"], ascending=False)

# Clean up output
#popularItemsClean = popularItems
popularItemsSorted["Item Price"] = popularItemsSorted["Item Price"].map("${:.2f}".format)
popularItemsSorted["Total Purchase Value"] = popularItemsSorted["Total Purchase Value"].map("${:.2f}".format)

popularItemsSorted.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
# Sort the Most Populat Items data frame by Total Purchase Value in descending order
mostProfitableItems = popularItems.sort_values(by=["Total Purchase Value"], ascending=False)

# Clean up output
mostProfitableItems["Item Price"] = mostProfitableItems["Item Price"].map("${:.2f}".format)
mostProfitableItems["Total Purchase Value"] = mostProfitableItems["Total Purchase Value"].map("${:.2f}".format)

mostProfitableItems.head()
```
